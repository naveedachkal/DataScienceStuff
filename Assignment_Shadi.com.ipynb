```python
#Import Libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#Read data (Here, products.csv file)
DataFrame = pd.read_csv("Products.csv", engine="python");
DataFrame
```

Working with Missing values

```python
#Check for NULL values
print(DataFrame.isnull())
```

```python
# As "Age" column is almost emplty so drop the complete column
Products_Details = DataFrame.drop('Age ', axis=1)
Products_Details
```

```python
#Now removing the rows where Case_cost value is missing
Products_Details = Products_Details[Products_Details.Case_cost.notnull()]
print(Products_Details.isnull().sum())
```

```python
#Making the datatype homogeneous 
Products_Details.iloc[:, -2].astype(float)
```

```python
#Datatype of the data
print(Products_Details.dtypes)
```

```python
#Distribution of Case Cost - to understand the Median Case Cost
Case_cost_Details = Products_Details.iloc[:, -1]
plt.hist(Case_cost_Details, bins = "auto", alpha = 0.5)
```

As abowe graph suggests, the Median Case cost is around 100$.

```python
#Distribution of Bottle Price - to understand the median Bottle Price
Bottle_Price_Details = Products_Details.iloc[:, -3]
plt.hist(Bottle_Price_Details, bins = "auto", alpha = 0.5)
```

As abowe graph suggests, the median price of the bottle is around 10$.

```python
# Bottle_size vs Bottle_Price
Bottle_Size = Products_Details.iloc[:, -3]
Bottle_Price = Products_Details.iloc[:, 6]
plt.scatter(Bottle_Size, Bottle_Price)
```

As abowe graph suggests, size & the price of the bottle are not necessarily proportional.
